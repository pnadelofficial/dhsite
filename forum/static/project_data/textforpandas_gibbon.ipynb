```python
import os
rootdir = '/content/FallDHCourseMaterials/Gibbon_By_Chapter_zipped'
df_list = []

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
      if file[-1] == 'u':
        if '_' in file[15:17]:
          conllu_df = tp.io.conll.conll_u_to_dataframes(os.path.join(subdir, file))[0]
          conllu_df['chapter'] = conllu_df['lemma'].apply(lambda x: int(file[15:16]))
          df_list.append(conllu_df)
        else:
          conllu_df = tp.io.conll.conll_u_to_dataframes(os.path.join(subdir, file))[0]
          conllu_df['chapter'] = conllu_df['lemma'].apply(lambda x: int(file[15:17]))
          df_list.append(conllu_df)

gibbon = pd.concat(df_list).sort_values('chapter')
```
